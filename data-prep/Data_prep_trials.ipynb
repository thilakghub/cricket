```python
batsman_data = master_odi_t20.groupby(['batsman', 'match_date','match_type','innings_number'])\
                            .apply(lambda x: pd.Series(
                                    dict(runs_batsman = (x.runs_batsman).sum()
                                         ,balls_faced = (x.ball).count()
                                         ,boundary_4 = (x.runs_batsman == 4).sum()
                                         ,boundary_6 = (x.runs_batsman == 6).sum()
                                         ,boundary_6 = (x.runs_batsman == 6).sum()
                                        ))).reset_index()
```

```python
batsman_data['strike_rate'] = (batsman_data['runs_batsman'] / batsman_data['balls_faced'] * 100)
```

```python
batsman_data = batsman_data.sort_values(['batsman', 'match_date','match_type'], ascending=[True, True, True])
```

```python
#standard batting points
batsman_data['batting_points'] = (batsman_data['runs_batsman'] * 1) \
                                    + (batsman_data['boundary_4'] * 1) \
                                    + (batsman_data['boundary_6'] * 2)

#conditional batting points based on runs
batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'T20')
                                          & (batsman_data['runs_batsman'] >= 50)
                                          , batsman_data['batting_points'] + 8
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'T20')
                                          & (batsman_data['runs_batsman'] >= 100)
                                          , batsman_data['batting_points'] + 16
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'ODI')
                                          & (batsman_data['runs_batsman'] >= 50)
                                          , batsman_data['batting_points'] + 4
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'ODI')
                                          & (batsman_data['runs_batsman'] >= 100)
                                          , batsman_data['batting_points'] + 8
                                          , batsman_data['batting_points'])

#conditional batting points based on strike rate
batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'T20')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] < 50)
                                          , batsman_data['batting_points'] - 6
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'T20')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] >= 50)
                                          & (batsman_data['strike_rate'] < 60)
                                          , batsman_data['batting_points'] - 4
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'T20')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] >= 60)
                                          & (batsman_data['strike_rate'] <= 70)
                                          , batsman_data['batting_points'] - 2
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'ODI')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] < 40)
                                          , batsman_data['batting_points'] - 6
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'ODI')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] >= 40)
                                          & (batsman_data['strike_rate'] < 50)
                                          , batsman_data['batting_points'] - 4
                                          , batsman_data['batting_points'])

batsman_data['batting_points'] = np.where((batsman_data['match_type'] == 'ODI')
                                          & (batsman_data['balls_faced'] >= 10)
                                          & (batsman_data['strike_rate'] >= 50)
                                          & (batsman_data['strike_rate'] <= 60)
                                          , batsman_data['batting_points'] - 2
                                          , batsman_data['batting_points'])
```
