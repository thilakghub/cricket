```python
batsman_data = master_odi_t20.groupby(['batsman', 'match_date'])\
                            .apply(lambda x: pd.Series(
                                    dict(runs_batsman = (x.runs_batsman).sum()
                                         ,balls_faced = (x.ball).count()
                                         ,boundary_4 = (x.runs_batsman == 4).sum()
                                         ,boundary_6 = (x.runs_batsman == 6).sum()
                                        ))).reset_index()
```

```python
batsman_data = batsman_data.sort_values(['batsman', 'match_date'], ascending=[True, True])
```
