#### moving average

we take an average of last 8 matches. to account for Nan we colasce it with last 4 matches

```python
batsman_data['moving_average_8'] = batsman_data.groupby(['batsman','match_type'])['runs_batsman'].rolling(8).mean().shift(1).reset_index()['runs_batsman']
batsman_data['moving_average_4'] = batsman_data.groupby(['batsman','match_type'])['runs_batsman'].rolling(4).mean().shift(1).reset_index()['runs_batsman']
batsman_data['moving_average'] = batsman_data.moving_average_8.combine_first(batsman_data.moving_average_4)
batsman_data = batsman_data.drop(['moving_average_8','moving_average_4'], axis = 1)
```

#### batting pitch or bowling pitch

```python
venue_details_innings_1 = master_t20[master_t20['innings_number'] == 1]\
                                        .groupby(['venue','match_type'])\
                                        .agg({"runs_batsman": np.sum, "match_date": pd.Series.nunique})\
                                        .reset_index()
```

```python
venue_details_innings_2 = master_t20[master_t20['innings_number'] == 2]\
                                        .groupby(['venue','match_type'])\
                                        .agg({"runs_batsman": np.sum, "match_date": pd.Series.nunique})\
                                        .reset_index()
```

```python
venue_details_innings = pd.merge(venue_details_innings_1, venue_details_innings_2
                                 ,how='left', on=['venue','match_type']
                                 ,suffixes=('_1', '_2'))
```

```python
venue_details_innings['avg_runs_innings_1'] = round(venue_details_innings['runs_batsman_1'] / venue_details_innings['match_date_1'])
venue_details_innings['avg_runs_innings_2'] = round(venue_details_innings['runs_batsman_2'] / venue_details_innings['match_date_2'])
```

```python
venue_details_innings.head()
```

```python
batsman_data = pd.merge(batsman_data, venue_details_innings[['venue','match_type','avg_runs_innings_1','avg_runs_innings_2']]
                                 ,how='left', on=['venue','match_type'])
```
