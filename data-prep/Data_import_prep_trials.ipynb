```python
ball_by_ball_data = pd.DataFrame(columns=columns)
```

```python
for j in range(len(df_yaml['innings'])):
    ball_by_ball_innings = pd.DataFrame(columns=columns)
    for i in range(len(list(list(df_yaml['innings'][j].values())[0]['deliveries']))):
        ball_by_ball = pd.DataFrame(columns=columns)
        ball_by_ball['innings'] = list(df_yaml['innings'][j].items())[0][0]
        ball_by_ball['team'] = list(df_yaml['innings'][j].values())[0]['team']

        ball_by_ball['ball'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][0]
        ball_by_ball['batsman'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['batsman']
        ball_by_ball['bowler'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['bowler']
        ball_by_ball['non_striker'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['non_striker']
        ball_by_ball['runs.batsman'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['runs']['batsman']
        ball_by_ball['runs.extras'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['runs']['extras']
        ball_by_ball['runs.total'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['runs']['total']
        ball_by_ball['wicket.kind'] = 'NA'
        ball_by_ball['wicket.player_out'] = 'NA'
        ball_by_ball['wicket.fielders'] = 'NA'

        if 'wicket' in list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]:
            ball_by_ball['wicket.kind'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['wicket']['kind']
            ball_by_ball['wicket.player_out'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['wicket']['player_out']
            ball_by_ball['wicket.fielders'] = 'NA'

            if 'fielders' in list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['wicket']:
                ball_by_ball['wicket.fielders'] = list(list(df_yaml['innings'][j].values())[0]['deliveries'][i].items())[0][1]['wicket']['fielders']
            else:
                ball_by_ball['wicket.fielders'] = 'NA'
        else:
            ball_by_ball['wicket.kind'] = 'NA'
            ball_by_ball['wicket.player_out'] = 'NA'
            ball_by_ball['wicket.fielders'] = 'NA'
        
        ball_by_ball_innings = ball_by_ball_innings.append(ball_by_ball)
    ball_by_ball_data = ball_by_ball_data.append(ball_by_ball_innings)
```

```python
ball_by_ball_data['match_date'] = "{:%Y-%m-%d}".format(df_yaml['info']['dates'][0])
ball_by_ball_data['match_type'] = df_yaml['info']['match_type']
ball_by_ball_data['venue'] = df_yaml['info']['venue']
ball_by_ball_data['city'] = df_yaml['info']['city']
ball_by_ball_data['overs'] = df_yaml['info']['overs']
ball_by_ball_data['winner'] = df_yaml['info']['outcome']['winner']
```
